**Task - 3 - Outlier Detection:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
#IQR for Score and GDP per Capita
Q1_score = south_asian_data['score'].quantile(0.25)
Q3_score = south_asian_data['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = south_asian_data['Log GDP per capita'].quantile(0.25)
Q3_gdp = south_asian_data['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Calculating the lower and upper bounds for Score and GDP per Capita
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers_score = south_asian_data[(south_asian_data['score'] < lower_bound_score) | (south_asian_data['score'] > upper_bound_score)]
outliers_gdp = south_asian_data[(south_asian_data['Log GDP per capita'] < lower_bound_gdp) | (south_asian_data['Log GDP per capita'] > upper_bound_gdp)]

# Displaying outliers
print("Outlier countries based on Score:")
print(outliers_score[['Country name', 'score']])

print("\nOutlier countries based on GDP per Capita:")
print(outliers_gdp[['Country name', 'Log GDP per capita']])
```

2. Define outliers using the 1.5 × IQR rule.

Ans: Outliers are values that stand out or are abnormal as compared to the overall pattern in a data. The 1.5 x IQR rule is a method to detect outliers in a dataset. In this method, Interquartile Range(IQR) is calculated by subtracting Q1 from Q3. Then, lower bound and upper bound are identified using the formula Q1-1.5*IQR and Q3+1.5*IQR, respectively.Hence, the values that lie below the lower bound or above the upper bound are outliers.

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python

def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

# Outliers for South Asia and Middle East
south_asia_outliers = detect_outliers(south_asian_data, 'score')
middle_east_outliers = detect_outliers(middle_east_data, 'score')

print("South Asia Outliers:")
print(south_asia_outliers[['Country name', 'score']])

print("Middle East Outliers:")
print(middle_east_outliers[['Country name', 'score']])

# Scatter plots for both regions
plt.figure(figsize=(12, 6))

# South Asia
plt.subplot(1, 2, 1)
plt.scatter(south_asian_data['Log GDP per capita'], south_asian_data['score'], color='green', label='South Asia')
plt.scatter(south_asia_outliers['Log GDP per capita'], south_asia_outliers['score'], color='red', label='Outliers')
plt.title('South Asia: GDP vs Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()

# Middle East
plt.subplot(1, 2, 2)
plt.scatter(middle_east_data['Log GDP per capita'], middle_east_data['score'], color='blue', label='Middle East')
plt.scatter(middle_east_outliers['Log GDP per capita'], middle_east_outliers['score'], color='red', label='Outliers')
plt.title('Middle East: GDP vs Score')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()

plt.tight_layout()
plt.show()
```
